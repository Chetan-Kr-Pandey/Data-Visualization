## Data Process (Cleaning)

```python
df_canada['Total']=df_canada.iloc[:,-34:].sum(axis=1)
df_canada.head()
```

```python
df_canada.set_index('Countries',inplace=True)
df_canada.head()
```

```python
df=df_canada.copy()
```

```python
df.tail()
```

```python
df.shape
```

```python
dff=df.copy()
dff.drop("Total",inplace=True)
```

```python
dff.tail()
```

## Line Plot

```python
years= list(map(str,range(1980,2014)))

df.loc["India",years].plot(kind="line",color="green",marker="+")
plt.title("Immigration from Indian to Canada in the year(1980-2013)")
plt.xlabel("Years")
plt.ylabel("Indians")
plt.show()
```

## Box Plot

## below 4 lines od codes are for designing the box plot

```python
boxprops = dict(linewidth=2, color='blue', facecolor='lightgray')  # Box properties
medianprops = dict(linestyle='-', linewidth=2, color='red')  # Median line properties
whiskerprops = dict(linestyle='--', linewidth=1, color='green')  # Whisker properties
capprops = dict(linewidth=2, color='purple')# Caps properties
```

```python
years= list(map(str,range(1980,2014)))
df_japan = df.loc['Japan',years].transpose()
df_japan.plot(kind='box' ,boxprops=boxprops, medianprops=medianprops,
                whiskerprops=whiskerprops, capprops=capprops, patch_artist=True)

plt.title('Box plot of Immigrants of Japanese from 1980-2013')
plt.ylabel('No. of Immigrants')
plt.show()
```

## Scatter plot

## some data processing here to create new dataframe which will only contain years and their total resp.

```python
years=dff.columns[5:39]

totals=df.iloc[:,5:39].sum(axis=0)

new_df=pd.DataFrame({'Years':years,'Total':totals})

new_df = new_df.reset_index(drop=True)

new_df.head()
```

```python
new_df.plot(kind="scatter",x='Years',y='Total' ,marker="*")

plt.gca().set_facecolor('pink')

plt.gcf().set_size_inches(8, 6)

plt.title('total Immigrant population to canada from 1980-2013')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('No. of Immigrants')

plt.show()
```

## Adding a Marker

```python
canada_map=folium.Map(
    location=[56.130,-106.35],
    zoom_start=4,
)
## add a red mark to Ontario

# create a feature group

ontario = folium.map.FeatureGroup()

# style the feature group
ontario.add_child(
    folium.features.CircleMarker(
        [51.25,-85.32], radius = 5,
        color="red", fill_color= "red")
)
canada_map.add_child(ontario)

canada_map
```

## for choropleth map we need countries and thier total immmigration. for that we do some data processing and create new data frame which will have these two columns

```python
df_cana=df_canada.copy()
df_cana.reset_index(level=0, inplace=True)
df_can=df_cana.drop(df_cana.iloc[:,1:40],axis=1)
df_can.head()
```

```python
df_can=df_can.drop(195)
df_can.tail()
#df_can['Total'].max()
#df_can['Total'].min()
```

```python
world_map=folium.Map(zoom_start=2)

## geojson file
world_geo= r'world.json'

world_map.choropleth(
    geo_data=world_geo,
    data=df_can,
    columns=['Countries','Total'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Immigration to Canada'
)
world_map
```
